```python
from google.colab import files

# Upload file from your computer
uploaded = files.upload()

# Get uploaded filename
input_file = list(uploaded.keys())[0]
print(f"Uploaded file: {input_file}")

# Split configuration
num_files = 10
total_rows = 640000
rows_per_file = total_rows // num_files  # 64,000

# Read file
with open(input_file, "r") as f:
    lines = f.readlines()

# Safety check
assert len(lines) == total_rows, \
    f"Expected {total_rows} rows, found {len(lines)}"

# Split and write files
for i in range(num_files):
    start = i * rows_per_file
    end = start + rows_per_file
    output_file = f"/content/output_part_{i+1}.txt"

    with open(output_file, "w") as f:
        f.writelines(lines[start:end])

    print(f"Created {output_file} with {rows_per_file} rows")
```
